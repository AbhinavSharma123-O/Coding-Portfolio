```python
import re

text = "Hello!!! This is an Example, with Numbers 123 and Symbols @#."

text = text.lower()
text = re.sub(r'[^a-z\s]', '', text)
text = re.sub(r'\s+', ' ', text).strip()

print("Cleaned Text:")
print(text)
```

```python
tokens = text.split()
print("Tokens:")
print(tokens)
```

```python
stop_words = {"is", "am", "are", "the", "this", "and", "a", "an", "with", "to", "of", "in"}

filtered_tokens = [word for word in tokens if word not in stop_words]

print("After Stop Word Removal:")
print(filtered_tokens)
```

```python
def rule_based_stem(word):
    if word.endswith("ing"):
        return word[:-3]
    elif word.endswith("ed"):
        return word[:-2]
    elif word.endswith("s"):
        return word[:-1]
    return word

stemmed_words = [rule_based_stem(word) for word in filtered_tokens]

print("Stemmed Words:")
print(stemmed_words)
```
