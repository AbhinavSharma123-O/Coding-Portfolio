```python
Median = dframe.median()
```

```python

medians = {}
for column_name in dframe.columns:
    data = dframe[column_name].values
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 1:
         median = sorted_data[n // 2]
    else:
        mid1 = sorted_data[n // 2 - 1]
        mid2 = sorted_data[n // 2]
        median = (mid1 + mid2) / 2
    medians[column_name] =  median
medians_series = ram.Series( medians)
print( medians_series)
```

```python
print('Manual Max Value Calculation for All Columns:')
max_values = {}
for column_name in dframe.columns:
    data = dframe[column_name].values

    # Manual calculation of maximum value
    if len(data) > 0:
        current_max = data[0]
        for value in data:
            if value > current_max:
                current_max = value
    else:
        current_max = None

    max_values[column_name] = current_max
max_values = ram.Series(max_values)

print("\nMax Values:")
print(max_values)
```
