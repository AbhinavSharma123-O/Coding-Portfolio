```python
import matplotlib.pyplot as plt
```

```python
age=[58,69,43,39,63,52,47,31,79,36]
```

```python
chol=[189,235,193,177,154,191,213,165,198,181]
```

```python
n=len(age)
sum_age=sum(age)
sum_chol=sum(chol)
```

```python
mean_age=sum_age/n
mean_chol=sum_chol/n
```

```python
sq_age=[]
sq_chol=[]
sum_agechole=0
```

```python
for i in age:
  sq_age.append(i**2)

```

```python
for j in range(len(age)):
  sum_agechole=sum_agechole+age[j]*chol[j]
```

```python
sum_sq_age = sum(sq_age)
b1 = (n * sum_agechole - sum_age * sum_chol) / (n * sum_sq_age - sum_age**2)
```

```python
b0=mean_chol-b1*mean_age
```

```python
print(b0+b1*60)
```

```python
plt.scatter(age,chol)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.title("Age vs Cholesterol")
plt.show()
```

```python
input_age = 60 # You can change this age to any value you want to predict for
predicted_cholesterol = b0 + b1 * input_age
print(f"For an age of {input_age}, the predicted cholesterol is: {predicted_cholesterol:.2f}")
```

```python
input_cholesterol = 200 # You can change this cholesterol to any value you want to predict for
predicted_age = (input_cholesterol - b0) / b1
print(f"For a cholesterol of {input_cholesterol}, the predicted age is: {predicted_age:.2f}")
```

```python
import random

# Create an empty list for weights
random_weights = []

# Add 10 random weights (e.g., between 50 and 100 kg)
for _ in range(n):
    random_weights.append(random.randint(50, 100))

print("Random Weights (kg):")
print(random_weights)
```

```python

sum_weights = sum(random_weights)
mean_weights = sum_weights / n
sum_weight_chol = 0
for i in range(n):
    sum_weight_chol += random_weights[i] * chol[i]
sq_weights = []
for i in random_weights:
    sq_weights.append(i**2)
sum_sq_weights = sum(sq_weights)
b1_weight_chol = (n * sum_weight_chol - sum_weights * sum_chol) / (n * sum_sq_weights - sum_weights**2)
b0_weight_chol = mean_chol - b1_weight_chol * mean_weights
print(f"b0 (intercept) for Weight vs Cholesterol: {b0_weight_chol:.2f}")
print(f"b1 (slope) for Weight vs Cholesterol: {b1_weight_chol:.2f}")
```

```python
input_weight = 70 # Change this weight to predict cholesterol
predicted_cholesterol_from_weight = b0_weight_chol + b1_weight_chol * input_weight

print(f"For a weight of {input_weight} kg, the predicted cholesterol is: {predicted_cholesterol_from_weight:.2f}")
```

```python
plt.scatter(random_weights, chol)
plt.xlabel("Weight (kg)")
plt.ylabel("Cholesterol")
plt.title("Weight vs Cholesterol")
plt.show()
```
