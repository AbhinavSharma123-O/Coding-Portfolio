```python
import cv2
import numpy as np

image = cv2.imread("/content/drive/MyDrive/ML_Dataset/4k-Red-Sun-Illuminating-Japanese-Temple-Mountain-Scene.jpg")
if image is None:
    print("Error: Could not load the image. Please check the file path and ensure the image exists.")
else:
    print("Original Shape:", image.shape)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
img_256 = cv2.resize(gray, (256, 256))
img_224 = cv2.resize(gray, (224, 224))
```

```python
h, w = img_256.shape
cropped = img_256[10:h-10, 10:w-10]
```

```python
img_float = cropped.astype(np.float32)
```

```python
img_minmax = img_float / 255.0
```

```python
mean = np.mean(img_float)
std = np.std(img_float)
img_zscore = (img_float - mean) / std
```

```python
mask = np.zeros_like(img_256, dtype=np.uint8)
center = (128, 128)
radius = 80
cv2.circle(mask, center, radius, 255, -1)
```

```python
masked_img = cv2.bitwise_and(img_256, img_256, mask=mask)
```

```python
plt.imshow(masked_img, cmap='gray')
plt.title("Circular Masked Image")
plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Original image (already loaded as 'image')
color_image = image

# Grayscale image (already created as 'gray')
grayscale_image = gray

# Black and white (binary) image
# We'll use a simple threshold. First, ensure it's grayscale if not already.
# Since 'gray' is already grayscale, we'll use that.
_, black_white_image = cv2.threshold(grayscale_image, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(15, 5))

# Display Original Color Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
plt.title('')
plt.axis('off')

# Display Grayscale Image
plt.subplot(1, 3, 2)
plt.imshow(grayscale_image, cmap='gray')
plt.title('')
plt.axis('off')

# Display Black and White Image
plt.subplot(1, 3, 3)
plt.imshow(black_white_image, cmap='gray')
plt.title('')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
h, w, _ = image.shape
h_half, w_half = h // 2, w // 2

top_left = image[0:h_half, 0:w_half]
top_right = image[0:h_half, w_half:w]
bottom_left = image[h_half:h, 0:w_half]
bottom_right = image[h_half:h, w_half:w]

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(top_left, cv2.COLOR_BGR2RGB))
plt.title('')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(top_right, cv2.COLOR_BGR2RGB))
plt.title('')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(bottom_left, cv2.COLOR_BGR2RGB))
plt.title('')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(bottom_right, cv2.COLOR_BGR2RGB))
plt.title('')
plt.axis('off')

plt.tight_layout()
plt.show()
```
